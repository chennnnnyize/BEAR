```python
def grad(x,alpha):
  p0=2*x[0]-x[1]
  p1=-x[0]+6*x[1]
  t=np.array([p0,p1])
  return t
  
x=np.array([[2.0],[2.0]])
alpha=0.3
f=[]
t=[]
xlis=[]
ylis=[]
for i in range(30):
  res=grad(x,alpha)
  x-=alpha*res

  sol=x[0]**2-x[0]*x[1]+3*x[1]**2+5
  f.append(sol)
  t.append(i)
  xlis.append(x[0][0])
  ylis.append(x[1][0])
fig = plt.figure()

points = plt.axes()
points.plot(xlis,ylis)
points.set_title('Points')
points.set_xlabel('x1')
points.set_ylabel('x2')
fig = plt.figure()
fplot = plt.axes()
fplot.set_title('value of f')
fplot.set_xlabel('iteration')
fplot.set_ylabel('f')
fplot.plot(t,f)
```

First order necessary: 

If x is a local minimizer and f is $C^1$ in an open neighborhood of x, suppose$\nabla f(x)\neq0$ and let direction $p=-\nabla f(x) ,\ g(p)=f(x+tp)$. There exist a small enough $\epsilon > 0 $ that the absolute value of $\nabla g(0)$ is greater than the higher order terms in taylor expansions for g around p = 0. Thus we could say that $g(p) < g(0)+ \nabla g(0)p+ |\nabla g(0)p|$. Since we had p be the opposite sign, $g(p)<g(0)$. However, it contradicts with the fact that $g(0)$ is the local minimum. Thus $\nabla f(x)=0$

Second order necesary:

Suppose that $g(p)=f(x+tp)$ and$p^\top \nabla^2 f(x)p <0$, there exists a small enough $\epsilon > 0 $ such that $p^\top \nabla^2 f(x+tp)p <0$. Taylor expansion of $$g(p) = g(0)+\nabla g(0)p+ p^\top \nabla^2 g(0)p+ higherorderterm(p^\top p)$$
. With the small enough $\epsilon>|p|$,$|higherorderterm(p^\top p)|<p^\top|\nabla^2 g(0)|p$.Thus we could say that $$g(p) < g(0)+ \nabla g(0)p+p^\top \nabla^2 g(0)p+ p^\top|\nabla^2 g(0)|p$$. With $p^\top \nabla^2 f(x)p <0$ and $\nabla f(x)=0$, we could reduce the function to $g(p)<g(0)$. However, it contradicts with the fact that $g(0)$ is the local minimum. Thus $\nabla f(x)=0$and $\nabla^2 f(x)\geq 0$.
